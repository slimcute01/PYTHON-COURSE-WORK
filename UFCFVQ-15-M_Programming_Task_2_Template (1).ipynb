### Requirement FR9 - Read CSV data from two files and merge it into a single Data Frame 

```python
# use an appropriate library
import pandas as pd

# read both csv files
data_frame1 = pd.read_csv("task2a.csv")
data_frame2 = pd.read_csv("task2b.csv")

# merge both files with the common column
merge_dataframe = data_frame1.merge(data_frame2, on = 'region')

merge_dataframe.info() # use info to confirm number of rolls and columns recieved
merge_dataframe # call desired result
```

### Requirement FR11 - Investigate the effects of household income on levels of bias crimes

### Requirement FR12 - Test the hypothesis that household income has no effect on levels of bias crime

```python
# import library
import scipy.stats

# assign the variables being tested
household_income = merge_dataframe['median_household_income']
biascrimes = merge_dataframe['avg_biascrimes_per_100k']

# pass variables through pearson function
R_value, P_value = scipy.stats.pearsonr(household_income, biascrimes)

# print results to 4 decimal places
print(f'The correlation coefficient R_value={R_value:.4f}, and the P-value = {P_value:.4f}\n')
```

*Ho: There is no relationship between the Household income and level of bias crimes*

*Hi: There is a relationship between the Household income and level of bias crimes* 

With a coefficient of 0.2958 our results is closer to 0 than 1. This confirm that there is a fairly weak relationship between the median household income and the level of biased crime. 

And with a P-Value of 0.0435 which is less than my significant level of (0.05) this samples contains adequate evidence to ***reject the Null Hypothesis Ho***.

We accept Hi and conclude that there is a statistically significant relationship between household income and the level of bias crime.

<p style="color:red; font-weight:bold; font-size:xx-small">MARK: 100%</p>
<p style="color:red; font-weight:bold; font-size:xx-small">FEEDBACK: Appropriate test statistic and good interpretation. </p>

### Requirement FR14 - Test the hypothesis that urbanisation has a significant effect on levels of bias crimes

```python
# assign variables being tested
metro_population = merge_dataframe['share_population_in_metro_areas']
biascrimes = merge_dataframe['avg_biascrimes_per_100k']

# pass assigned variables into pearson function
R_value, P_value = scipy.stats.pearsonr(metro_population, biascrimes)

# print desired result to 4 decimal places
print(f'The correlation coefficient R_value={R_value:.4f}, and the P-value = {P_value:.4f}\n')
```

*Hi: There is a relationship between the Metropolitan population and level of bias crimes* 

With a coefficient of 0.2095 our results is closer to 0 than 1. This confirm that there is a positive but fairly weak relationship between the median household income and the level of biased crime. 

But with a P-Value of 0.1576 which is greater than my significant level of (0.05) this samples contains adequate evidence to ***accept the Null Hypothesis*** and reject the Alternative Hypothesis.

We accept Ho and conclude that there is no statistically significant relationship between Metropolitan population and the level of bias crime.
